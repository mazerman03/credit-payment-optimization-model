Leemos las tablas y concatenamos los datos

```python
import pandas as pd

df1 = pd.read_csv(r"data\raw\ListaCobroDetalle2022.csv")

df2 = pd.read_csv(r"data\raw\ListaCobroDetalle2023.csv")

id_bancos = pd.read_csv(r"data\raw\CatBanco.csv")

id_emisoras = pd.read_csv(r"data\raw\ListaCobroEmisora.csv")
cat_emisoras = pd.read_csv(r"data\raw\CatEmisora.csv")

train_data = pd.concat([df1,df2], ignore_index=False)

print(train_data.head())
```

Agregamos el nombre del banco por ID

```python
train_data = train_data.merge(id_bancos, left_on="idBanco", right_on="IdBanco", how="left")

print(train_data.columns)
```

```python
# train_data = train_data.drop(columns=["IdBanco_x", "IdBanco_y", "Nombre_y"])
train_data = train_data.rename(columns={"Nombre_x":"Nombre_banco"})
train_data.head()
```

a√±adimos el id de la emisora

```python
train_data = train_data.merge(
 id_emisoras,
 on="idListaCobro",
 how="left"   
)
# train_data = train_data.drop(columns=["idEmisora_y"])
train_data = train_data.rename(columns={"idEmisora_x":"IdEmisora"})
train_data.head()
```

Llenamos los dem√°s datos de la emisora

```python
train_data = train_data.merge(
    cat_emisoras,
    on="idEmisora",
    how="left"
)
train_data = train_data.drop(columns=["IdBanco_x", "IdBanco_y"])#, "IdBanco", "Emisora_y", "TipoEnvio_y"])
train_data = train_data.rename(columns={"Nombre_x":"NombreBanco", "Nombre_y":"NombreEmisora", "Emisora":"EmisoraCategoria", "TipoEnvio_x":"TipoEnvio",})
train_data.head()
```

Checar si hay un monto cobrado menor al exigible

```python
cond_monto_mayor_0 = (train_data["montoCobrado"] > 0) #< (train_data["montoExigible"])
# train_data["Monto_cobrado_menor_exigible"] = "Mayor"
train_data["MontoCobradoMayorQueCero"] = False
# train_data.loc[cond, "Monto_cobrado_menor_exigible"] = "Menor"
train_data.loc[cond_monto_mayor_0, "MontoCobradoMayorQueCero"] = True
# train_data["Monto_cobrado_menor_exigible"].value_counts()
# cond_monto_mayor_0
train_data["MontoCobradoMayorQueCero"].value_counts()
MayorACero = train_data[train_data["MontoCobradoMayorQueCero"]]
# print(MayorACero)
cond_cobrado_menor_a_exigible = MayorACero["montoCobrado"] < MayorACero["montoExigible"]
MayorACero["CobradoMenorExigible"] = False
MayorACero.loc[cond_cobrado_menor_a_exigible, "CobradoMenorExigible"] = True
MayorACero["CobradoMenorExigible"].value_counts()
CobradoMenorExigible = MayorACero[MayorACero["CobradoMenorExigible"]]
CobradoMenorExigible
```

Parametros del Modelo

```python
Model_Params = train_data.groupby("idCredito")["montoExigible"].sum().reset_index()
train_data["MontoADeber"] = train_data["montoExigible"] - train_data["montoCobrado"]
Model_Params = train_data.groupby("idCredito")[["montoExigible", "MontoADeber"]].sum().reset_index()
Model_Params = Model_Params.rename(columns={"montoExigible":"MontoTotal"})
Model_Params
```
